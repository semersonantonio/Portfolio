### Gym Member Analytics and Churn Prediction

This project analyzes gym member behavior and builds a machine learning model to predict customer churn.

The analysis includes:

- Business metrics
- Behavioral analytics
- Customer retention analysis
- Revenue trends
- Churn prediction using machine learning

---

#### Dataset Overview

The analysis uses three datasets:

- members
- checkins
- payments

These datasets simulate real gym operations including member behavior and subscription payments.

---

#### Analysis Sections

##### 1. Data Preparation

- Dataset loading
- Datetime conversion
- Data quality checks
- Missing value inspection
- Duplicate detection

---

##### 2. Business Overview

Key metrics:

- Total members
- Active vs churned members
- Churn rate
- Plan distribution

---

##### 3. Behavioral Analysis

Member engagement metrics:

- Total check-ins
- Weekly visit frequency
- Engagement segmentation

---

##### 4. Usage Patterns

Operational insights:

- Peak hours
- Busiest weekdays

---

##### 5. Retention Analysis

Customer lifecycle metrics:

- Membership duration
- Churn rate
- Cohort retention analysis

---

##### 6. Revenue Analysis

Monthly revenue trends based on payment transactions.

---

##### 7. Machine Learning

A churn prediction model is trained using:

Features:

- age
- membership_days
- total_checkins
- avg_weekly_visits

Model used:

Random Forest Classifier

---

##### Model Evaluation

Metrics:

- Accuracy
- Precision

- ROC-AUC

---

#### Churn Risk Identification

The model is applied to all members to estimate churn probability and identify high-risk members.

This type of analysis can support:

- retention campaigns
- targeted engagement programs
- proactive customer management

```python
# ============================================
# ENVIRONMENT SETUP
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve
from sklearn.metrics import classification_report, roc_auc_score

import joblib
```

```python
# ----------------------------------------------------------
# Visualization Settings
# ----------------------------------------------------------
# Configure the notebook display and plotting style.
#
# display.max_columns
#     Ensures that pandas displays all dataframe columns.
#
# seaborn style
#     Applies a cleaner and more readable plot style.
# ----------------------------------------------------------

pd.set_option("display.max_columns", None)
sns.set_style("whitegrid")

print("Environment successfully configured.")
```

```python
# ============================================
# LOAD DATASETS
# ============================================

# Load the datasets used in this analysis.
# Each dataset represents a different aspect
# of gym operations.

# members: demographic and membership information
# checkins: gym attendance history
# payments: payment transactions

members = pd.read_csv("data/members.csv")
checkins = pd.read_csv("data/checkins.csv")
payments = pd.read_csv("data/payments.csv")

print("Datasets successfully loaded.\n")

print("Members shape:", members.shape)
print("Checkins shape:", checkins.shape)
print("Payments shape:", payments.shape)
```

```python
# ============================================
# DATETIME CONVERSION AND INITIAL INSPECTION
# ============================================

# Convert date columns to datetime format
# This allows proper time-based analysis.

members["join_date"] = pd.to_datetime(members["join_date"])
members["cancel_date"] = pd.to_datetime(members["cancel_date"])

checkins["date"] = pd.to_datetime(checkins["date"])
payments["date"] = pd.to_datetime(payments["date"])

print("Datetime conversion completed.\n")
```

```python
# --------------------------------------------
# Preview the datasets
# --------------------------------------------

print("Members preview:")
display(members.head())

print("\nCheckins preview:")
display(checkins.head())

print("\nPayments preview:")
display(payments.head())
```

```python
# --------------------------------------------
# Check data types
# --------------------------------------------

print("\nData types - MEMBERS")
print(members.dtypes)

print("\nData types - CHECKINS")
print(checkins.dtypes)

print("\nData types - PAYMENTS")
print(payments.dtypes)
```

```python
# ============================================
# DATA QUALITY CHECK
# ============================================

# In this section we inspect the datasets to identify:
# - Missing values
# - Duplicate records
# - Basic descriptive statistics


# --------------------------------------------
# Missing values
# --------------------------------------------

print("==== MISSING VALUES CHECK ====\n")

print("Members dataset:")
print(members.isnull().sum())

print("\nCheckins dataset:")
print(checkins.isnull().sum())

print("\nPayments dataset:")
print(payments.isnull().sum())
```

```python
# --------------------------------------------
# Duplicate records
# --------------------------------------------

print("\n==== DUPLICATE RECORDS ====\n")

print("Duplicates in members:", members.duplicated().sum())
print("Duplicates in checkins:", checkins.duplicated().sum())
print("Duplicates in payments:", payments.duplicated().sum())
```

```python
# --------------------------------------------
# Descriptive statistics
# --------------------------------------------

print("\n==== DESCRIPTIVE STATISTICS ====\n")

print("Members statistics:")
display(members.describe())

print("\nPayments statistics:")
display(payments.describe())
```

```python
# ============================================
# BUSINESS OVERVIEW
# ============================================

# This section provides a high-level overview of the gym business.
# We calculate basic metrics such as:
# - Total number of members
# - Active vs churned members
# - Churn rate
# - Membership plan distribution


print("===== GYM BUSINESS OVERVIEW =====\n")


# --------------------------------------------
# Total number of members
# --------------------------------------------

total_members = len(members)

print("Total number of members:", total_members)


# --------------------------------------------
# Active vs churned members
# --------------------------------------------

members["is_active"] = members["cancel_date"].isna()

status_counts = members["is_active"].value_counts()

print("\nActive vs churned members:")
print(status_counts)


# --------------------------------------------
# Churn rate
# --------------------------------------------

churn_rate = members["cancel_date"].notna().mean()

print("\nOverall churn rate:", round(churn_rate * 100, 2), "%")


# --------------------------------------------
# Membership plan distribution
# --------------------------------------------

print("\nMembership plan distribution:")

print(members["plan_type"].value_counts())
```

```python
# ============================================
# OVERVIEW VISUALIZATIONS
# ============================================


# --------------------------------------------
# Active vs churned members
# --------------------------------------------

status_counts = members["is_active"].value_counts()

status_counts.plot(kind="bar")

plt.title("Active vs Churned Members")
plt.xlabel("Member Status")
plt.ylabel("Number of Members")

plt.xticks([0, 1], ["Active", "Churned"])

plt.show()
```

```python
# --------------------------------------------
# Membership plan distribution
# --------------------------------------------

members["plan_type"].value_counts().plot(kind="bar")

plt.title("Membership Plan Distribution")
plt.xlabel("Plan Type")
plt.ylabel("Number of Members")

plt.show()
```

```python
# ============================================
# GYM GROWTH ANALYSIS
# ============================================

# In this section we analyze the growth of the gym
# by calculating how many new members joined each year.


# --------------------------------------------
# Extract join year
# --------------------------------------------

members["join_year"] = members["join_date"].dt.year
```

```python
# --------------------------------------------
# Count new members per year
# --------------------------------------------

members_per_year = members.groupby("join_year").size()

print("New members per year:")
print(members_per_year)
```

```python
# --------------------------------------------
# Growth visualization
# --------------------------------------------

members_per_year.plot(kind="bar")

plt.title("New Members per Year")
plt.xlabel("Year")
plt.ylabel("Number of New Members")

plt.show()
```

```python
# ============================================
# AGE ANALYSIS
# ============================================

# This section analyzes the age distribution of gym members
# to better understand the demographic profile.


print("===== AGE ANALYSIS =====\n")


# --------------------------------------------
# Basic age statistics
# --------------------------------------------

print("Age statistics:")

display(members["age"].describe())
```

```python
# --------------------------------------------
# Age distribution visualization
# --------------------------------------------

plt.figure()

members["age"].hist(bins=15)

plt.title("Age Distribution of Gym Members")
plt.xlabel("Age")
plt.ylabel("Number of Members")

plt.show()
```

```python
# ============================================
# TOTAL CHECK-INS PER MEMBER
# ============================================

# This section measures member engagement by calculating
# how many times each member visited the gym.


# --------------------------------------------
# Count check-ins per member
# --------------------------------------------

checkins_per_member = checkins.groupby("member_id").size()


# --------------------------------------------
# Add check-in count to members dataset
# --------------------------------------------

members["total_checkins"] = members["member_id"].map(checkins_per_member)


# Members with no check-ins receive 0
members["total_checkins"] = members["total_checkins"].fillna(0)
```

```python
# --------------------------------------------
# Check-in statistics
# --------------------------------------------

print("===== CHECK-INS PER MEMBER =====\n")

display(members["total_checkins"].describe())
```

```python
# --------------------------------------------
# Visualization
# --------------------------------------------

plt.figure()

members["total_checkins"].hist(bins=20)

plt.title("Distribution of Check-ins per Member")
plt.xlabel("Number of Check-ins")
plt.ylabel("Number of Members")

plt.show()
```

```python
# ============================================
# MEMBERSHIP DURATION
# ============================================

# This section calculates how long each member
# stayed at the gym.


# --------------------------------------------
# Define simulation end date
# --------------------------------------------

# The dataset was generated with a final
# reference date for active members.

END_DATE = pd.Timestamp("2025-01-01")


# --------------------------------------------
# Define each member's end date
# --------------------------------------------

# If the member canceled, use cancel_date.
# Otherwise use the simulation end date.

members["end_date"] = members["cancel_date"].fillna(END_DATE)


# --------------------------------------------
# Calculate membership duration in days
# --------------------------------------------

members["membership_days"] = (
    members["end_date"] - members["join_date"]
).dt.days
```

```python
# --------------------------------------------
# Membership duration statistics
# --------------------------------------------

print("===== MEMBERSHIP DURATION (DAYS) =====\n")

display(members["membership_days"].describe())
```

```python
# --------------------------------------------
# Convert duration to months (more interpretable)
# --------------------------------------------

members["membership_months"] = members["membership_days"] / 30
```

```python
# --------------------------------------------
# Visualization
# --------------------------------------------

plt.figure()

members["membership_months"].hist(bins=20)

plt.title("Membership Duration Distribution")
plt.xlabel("Months as Member")
plt.ylabel("Number of Members")
```

```python
# ============================================
# AVERAGE WEEKLY VISITS
# ============================================

# This metric estimates how frequently members
# attend the gym on a weekly basis.


# --------------------------------------------
# Calculate weekly training frequency
# --------------------------------------------

members["avg_weekly_visits"] = (
    members["total_checkins"] /
    (members["membership_days"] / 7)
)


# --------------------------------------------
# Handle potential infinite or missing values
# --------------------------------------------

members["avg_weekly_visits"] = members["avg_weekly_visits"].replace(
    [np.inf, -np.inf], 0
).fillna(0)
```

```python
# --------------------------------------------
# Statistics
# --------------------------------------------

print("===== WEEKLY TRAINING FREQUENCY =====\n")

display(members["avg_weekly_visits"].describe())
```

```python
# --------------------------------------------
# Visualization
# --------------------------------------------

plt.figure()

members["avg_weekly_visits"].hist(bins=7)

plt.title("Weekly Gym Visits Distribution")
plt.xlabel("Visits per Week")
plt.ylabel("Number of Members")

plt.show()
```

```python
# ============================================
# MEMBER ENGAGEMENT SEGMENTATION
# ============================================

# Members are segmented according to their
# average weekly training frequency.


# --------------------------------------------
# Define engagement classification
# --------------------------------------------

def classify_member(visits):

    if visits < 1:
        return "Low engagement"

    elif visits < 3:
        return "Moderate engagement"

    else:
        return "High engagement"
```

```python
# --------------------------------------------
# Apply segmentation
# --------------------------------------------

members["engagement_segment"] = members["avg_weekly_visits"].apply(classify_member)
```

```python
# --------------------------------------------
# Segment distribution
# --------------------------------------------

print("Engagement segment distribution:")

print(members["engagement_segment"].value_counts())
```

```python
# --------------------------------------------
# Visualization
# --------------------------------------------

members["engagement_segment"].value_counts().plot(kind="bar")

plt.title("Member Engagement Segments")
plt.xlabel("Engagement Level")
plt.ylabel("Number of Members")

plt.show()
```

```python
# ============================================
# CHECK-INS BY HOUR
# ============================================

# This section analyzes gym usage patterns
# throughout the day by counting check-ins
# for each hour.


# --------------------------------------------
# Count check-ins per hour
# --------------------------------------------

checkins_per_hour = checkins["hour"].value_counts().sort_index()

print("Check-ins by hour:")

display(checkins_per_hour)
```

```python
# --------------------------------------------
# Visualization
# --------------------------------------------

plt.figure()

checkins_per_hour.plot(kind="bar")

plt.title("Gym Check-ins by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Check-ins")

plt.show()
```

```python
# ============================================
# CHECK-INS BY WEEKDAY
# ============================================

# This section analyzes which days of the week
# have the highest gym attendance.


# --------------------------------------------
# Extract weekday from check-in date
# --------------------------------------------

checkins["weekday"] = checkins["date"].dt.day_name()


# --------------------------------------------
# Count check-ins per weekday
# --------------------------------------------

checkins_per_day = checkins["weekday"].value_counts()

print("Check-ins by weekday:")

display(checkins_per_day)
```

```python
# --------------------------------------------
# Order weekdays correctly
# --------------------------------------------

ordered_days = [
    "Monday",
    "Tuesday",
    "Wednesday",
    "Thursday",
    "Friday",
    "Saturday",
    "Sunday"
]

checkins_per_day = checkins_per_day.reindex(ordered_days)
```

```python
# --------------------------------------------
# Visualization
# --------------------------------------------

plt.figure()

checkins_per_day.plot(kind="bar")

plt.title("Gym Check-ins by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Check-ins")

plt.show()
```

```python
# ============================================
# CHURN ANALYSIS
# ============================================

# In this section we analyze member churn
# and basic retention metrics.


# --------------------------------------------
# Create churn variable
# --------------------------------------------

members["churn"] = members["cancel_date"].notna().astype(int)

print("Churn distribution:")

display(members["churn"].value_counts())


# --------------------------------------------
# Calculate churn rate
# --------------------------------------------

churn_rate = members["churn"].mean()

print("\nOverall churn rate:", round(churn_rate * 100, 2), "%")


# --------------------------------------------
# Average membership duration
# --------------------------------------------

avg_membership = members["membership_months"].mean()

print("\nAverage membership duration:",
      round(avg_membership, 2),
      "months")
```

```python
# --------------------------------------------
# Membership duration visualization
# --------------------------------------------

plt.figure()

members["membership_months"].hist(bins=20)

plt.title("Member Retention Duration")
plt.xlabel("Months as Member")
plt.ylabel("Number of Members")

plt.show()
```

```python
# ============================================
# TRAINING FREQUENCY VS CHURN
# ============================================

# In this section we analyze the relationship between
# training frequency and churn.
# The goal is to understand whether members who train
# less frequently are more likely to cancel their membership.


plt.figure(figsize=(8,5))

sns.boxplot(
    x=members["churn"],
    y=members["avg_weekly_visits"]
)

plt.title("Weekly Training Frequency vs Churn")
plt.xlabel("Churn (0 = Active Member, 1 = Cancelled Member)")
plt.ylabel("Average Weekly Visits")

plt.show()
```

```python
# ============================================
# REVENUE ANALYSIS
# ============================================

# This section analyzes the gym's revenue over time
# using the payments dataset.


# --------------------------------------------
# Create year-month column
# --------------------------------------------

payments["year_month"] = payments["date"].dt.to_period("M")


# --------------------------------------------
# Calculate monthly revenue
# --------------------------------------------

monthly_revenue = payments.groupby("year_month")["amount"].sum()

print("Monthly revenue:")

display(monthly_revenue.head())
```

```python
# --------------------------------------------
# Visualization
# --------------------------------------------

plt.figure(figsize=(10,5))

monthly_revenue.plot()

plt.title("Gym Monthly Revenue")
plt.xlabel("Time")
plt.ylabel("Revenue")

plt.show()
```

```python
# ============================================
# COHORT ANALYSIS — RETENTION BY JOIN COHORT
# ============================================

# Cohort analysis groups members by their join month
# and measures the average membership duration.


# --------------------------------------------
# Create join cohort
# --------------------------------------------

members["join_month"] = members["join_date"].dt.to_period("M")


# --------------------------------------------
# Calculate average membership duration by cohort
# --------------------------------------------

cohort = members.groupby("join_month")["membership_months"].mean()
```

```python
# --------------------------------------------
# Visualization
# --------------------------------------------

plt.figure()

cohort.plot()

plt.title("Average Membership Duration by Join Cohort")
plt.xlabel("Join Cohort")
plt.ylabel("Average Months as Member")

plt.show()
```

```python
# ============================================
# MACHINE LEARNING DATASET
# ============================================

# In this section we prepare the dataset that will
# be used to train the churn prediction model.

# The selected features represent member behavior
# and engagement patterns.


# --------------------------------------------
# Select model features
# --------------------------------------------

features = members[
    [
        "age",
        "membership_days",
        "total_checkins",
        "avg_weekly_visits"
    ]
]


# --------------------------------------------
# Target variable (churn)
# --------------------------------------------

target = members["churn"]


# --------------------------------------------
# Dataset verification
# --------------------------------------------

print("Features shape:", features.shape)
print("Target shape:", target.shape)

print("\nFeature sample:")

display(features.head())
```

```python
# ============================================
# TRAIN CHURN PREDICTION MODEL
# ============================================

# Split the dataset into training and testing sets.

X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.2,
    random_state=42,
    stratify=target
)

print("Training set:", X_train.shape)
print("Test set:", X_test.shape)
```

```python
# --------------------------------------------
# Initialize the model
# --------------------------------------------

model = RandomForestClassifier(
    n_estimators=300,
    max_depth=6,
    random_state=42
)
```

```python
# --------------------------------------------
# Train the model
# --------------------------------------------

model.fit(X_train, y_train)

print("\nModel successfully trained.")
```

```python
# ============================================
# MODEL EVALUATION
# ============================================

# --------------------------------------------
# Generate predictions
# --------------------------------------------

predictions = model.predict(X_test)
probabilities = model.predict_proba(X_test)[:,1]
```

```python
# --------------------------------------------
# Classification report
# --------------------------------------------

print("===== CLASSIFICATION REPORT =====\n")

print(classification_report(y_test, predictions))
```

```python
# --------------------------------------------
# ROC-AUC score
# --------------------------------------------

roc = roc_auc_score(y_test, probabilities)

print("\nROC-AUC:", round(roc,4))
```

```python
# --------------------------------------------
# Feature importance
# --------------------------------------------

importance = pd.DataFrame({
    "feature": features.columns,
    "importance": model.feature_importances_
}).sort_values("importance", ascending=False)

print("\n===== FEATURE IMPORTANCE =====\n")

display(importance)
```

```python
# -----------------------------
# Visualization
# -----------------------------

plt.figure()

importance.plot(
    x="feature",
    y="importance",
    kind="bar",
    legend=False
)

plt.title("Model Feature Importance")
plt.xlabel("Features")
plt.ylabel("Importance")

plt.show()
```

```python
# ============================================
# ROC CURVE
# ============================================

fpr, tpr, thresholds = roc_curve(y_test, probabilities)

plt.figure()

plt.plot(fpr, tpr)
plt.plot([0,1],[0,1], linestyle="--")

plt.title("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.show()
```

```python
# ============================================
# CHURN RISK IDENTIFICATION
# ============================================

# Calculate churn probability for all members

members["churn_probability"] = model.predict_proba(features)[:,1]
```

```python
# --------------------------------------------
# Preview probabilities
# --------------------------------------------

print("Example churn probabilities:\n")

display(
    members[
        [
            "member_id",
            "avg_weekly_visits",
            "total_checkins",
            "membership_months",
            "churn_probability"
        ]
    ].head()
)
```

```python
# --------------------------------------------
# Identify members with highest churn risk
# --------------------------------------------

high_risk_members = members.sort_values(
    "churn_probability",
    ascending=False
).head(20)


print("Top 20 members with highest churn risk:\n")

display(
    high_risk_members[
        [
            "member_id",
            "age",
            "avg_weekly_visits",
            "total_checkins",
            "membership_months",
            "churn_probability"
        ]
    ]
)
```

```python
# ============================================
# CHURN PROBABILITY DISTRIBUTION
# ============================================

plt.figure()

members["churn_probability"].hist(bins=20)

plt.title("Distribution of Churn Probability")
plt.xlabel("Churn Probability")
plt.ylabel("Number of Members")

plt.show()
```
