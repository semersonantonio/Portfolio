```python
# ============================================
# GYM DATASET GENERATION
# ============================================

import pandas as pd
import numpy as np
import random
from datetime import timedelta

np.random.seed(42)
random.seed(42)

pd.set_option("display.max_columns", None)

print("Ambiente configurado.")
```

```python
# ============================================
# DATASET PARAMETERS
# ============================================

N_MEMBERS = 350

START_DATE = pd.Timestamp("2018-01-01")
END_DATE = pd.Timestamp("2025-01-01")

PLAN_TYPES = [
    "monthly",
    "quarterly",
    "annual"
]

PLAN_PRICES = {
    "monthly": 120,
    "quarterly": 300,
    "annual": 1000
}

MIN_AGE = 18
MAX_AGE = 60

print("Parâmetros definidos.")
```

```python
# ============================================
# GENERATE MEMBERS
# ============================================

members_list = []

for i in range(N_MEMBERS):

    member_id = i + 1
    
    age = random.randint(MIN_AGE, MAX_AGE)
    
    plan_type = random.choice(PLAN_TYPES)
    
    join_date = START_DATE + pd.Timedelta(
        days=random.randint(0, (END_DATE - START_DATE).days)
    )
    
    engagement_level = random.choices(
        ["low", "medium", "high"],
        weights=[0.3, 0.5, 0.2]
    )[0]

    members_list.append({
        "member_id": member_id,
        "age": age,
        "plan_type": plan_type,
        "join_date": join_date,
        "engagement_level": engagement_level
    })

members = pd.DataFrame(members_list)

members.head()
```

```python
# ============================================
# GENERATE CHURN
# ============================================

cancel_dates = []

for _, row in members.iterrows():
    
    join_date = row["join_date"]
    engagement = row["engagement_level"]
    
    if engagement == "low":
        churn_prob = 0.6
    elif engagement == "medium":
        churn_prob = 0.35
    else:
        churn_prob = 0.15

    churn = np.random.rand() < churn_prob

    if churn:
        
        min_days = 60
        max_days = (END_DATE - join_date).days
        
        if max_days <= min_days:
            cancel_dates.append(pd.NaT)
            continue
        
        cancel_after = random.randint(min_days, max_days)
        cancel_date = join_date + pd.Timedelta(days=cancel_after)

        cancel_dates.append(cancel_date)
        
    else:
        cancel_dates.append(pd.NaT)

members["cancel_date"] = cancel_dates

members.head()
```

```python
# ============================================
# GENERATE CHECKINS
# ============================================

checkins_list = []

for _, member in members.iterrows():
    
    member_id = member["member_id"]
    join_date = member["join_date"]
    cancel_date = member["cancel_date"]
    engagement = member["engagement_level"]
    
    end_date = cancel_date if pd.notna(cancel_date) else END_DATE
    
    current_date = join_date
    
    while current_date < end_date:
        
        if engagement == "low":
            visits = random.randint(0,1)
        elif engagement == "medium":
            visits = random.randint(1,3)
        else:
            visits = random.randint(3,5)
        
        for _ in range(visits):
            
            day_offset = random.randint(0,6)
            visit_date = current_date + pd.Timedelta(days=day_offset)
            
            if visit_date < end_date:
                
                hour = random.randint(6,21)
                
                checkins_list.append({
                    "member_id": member_id,
                    "date": visit_date,
                    "hour": hour
                })
        
        current_date += pd.Timedelta(days=7)

checkins = pd.DataFrame(checkins_list)

checkins.head()
```

```python
# ============================================
# GENERATE PAYMENTS
# ============================================

payments_list = []

for _, member in members.iterrows():
    
    member_id = member["member_id"]
    join_date = member["join_date"]
    cancel_date = member["cancel_date"]
    plan = member["plan_type"]
    
    end_date = cancel_date if pd.notna(cancel_date) else END_DATE
    
    current_date = join_date
    
    while current_date < end_date:
        
        payment_amount = PLAN_PRICES[plan]
        
        payments_list.append({
            "member_id": member_id,
            "date": current_date,
            "amount": payment_amount,
            "plan_type": plan
        })
        
        if plan == "monthly":
            current_date += pd.Timedelta(days=30)
            
        elif plan == "quarterly":
            current_date += pd.Timedelta(days=90)
            
        else:
            current_date += pd.Timedelta(days=365)

payments = pd.DataFrame(payments_list)

payments.head()
```

```python
# ============================================
# VALIDATION
# ============================================

print("Members:", members.shape)
print("Checkins:", checkins.shape)
print("Payments:", payments.shape)

print("\nDuplicatas checkins:", checkins.duplicated().sum())
print("Duplicatas payments:", payments.duplicated().sum())

checkins = checkins.drop_duplicates()
payments = payments.drop_duplicates()

print("\nDataset validado.")
```

```python
# ============================================
# SAVE DATASETS
# ============================================

members.to_csv("members.csv", index=False)
checkins.to_csv("checkins.csv", index=False)
payments.to_csv("payments.csv", index=False)

print("Datasets salvos com sucesso.")
```
