```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report



members = pd.read_csv("members.csv")
checkins = pd.read_csv("checkins.csv")
payments = pd.read_csv("payments.csv")

members.head()
```

```python
members.info()
checkins.info()
payments.info()
```

```python
members["join_date"] = pd.to_datetime(members["join_date"])
members["cancel_date"] = pd.to_datetime(members["cancel_date"])

checkins["date"] = pd.to_datetime(checkins["date"])
payments["date"] = pd.to_datetime(payments["date"])
```

```python
members["churn"] = members["cancel_date"].notna().astype(int)
```

```python
members.head()
```

```python
# tempo de permanencia
members["membership_days"] = (
    members["cancel_date"].fillna(pd.Timestamp("2025-01-01"))
    - members["join_date"]
).dt.days
```

```python
# Frequencia de checkins
checkins_per_member = checkins.groupby("member_id").size()

members["total_checkins"] = members["member_id"].map(checkins_per_member)
members["total_checkins"] = members["total_checkins"].fillna(0)
```

```python
# Frequencia de media semanal
members["weeks_active"] = members["membership_days"] / 7

members["avg_weekly_visits"] = (
    members["total_checkins"] / members["weeks_active"]
)
```

```python
members.describe()
```

```python
# Tamanho da base
print("Número de alunos:", members.shape[0])
print("Número de checkins:", checkins.shape[0])
print("Número de pagamentos:", payments.shape[0])
```

```python
# Distribuicao de idades
plt.hist(members["age"], bins=20)
plt.title("Distribuição de Idade dos Alunos")
plt.xlabel("Idade")
plt.ylabel("Quantidade")
plt.show()
```

```python
# Distribuicao de generos
members["gender"].value_counts().plot(kind="bar")

plt.title("Distribuição por Gênero")
plt.ylabel("Número de Alunos")

plt.show()
```

```python
# Tipos de planos
members["plan"].value_counts().plot(kind="bar")

plt.title("Distribuição de Planos")
plt.ylabel("Número de Alunos")

plt.show()
```

```python
# Taxa de cancelamento
churn_rate = members["churn"].mean()

print("Taxa de churn:", round(churn_rate*100,2), "%")
```

```python
# Tempo de permanencia
plt.hist(members["membership_days"]/30, bins=20)

plt.title("Tempo de Permanência dos Alunos (meses)")
plt.xlabel("Meses")
plt.ylabel("Número de alunos")

plt.show()
```

```python
# Frequencia de media semanal
plt.hist(members["avg_weekly_visits"], bins=20)

plt.title("Frequência Média Semanal")
plt.xlabel("Visitas por semana")
plt.ylabel("Número de alunos")

plt.show()
```

```python
# Frequencia vs Cancelamento
import seaborn as sns

sns.boxplot(x="churn", y="avg_weekly_visits", data=members)

plt.title("Frequência de Treino vs Cancelamento")

plt.show()
```

```python
# Evolucao dos novos alunos por tempo
members["join_month"] = members["join_date"].dt.to_period("M")

members_per_month = members.groupby("join_month").size()

members_per_month.plot()

plt.title("Novos Alunos por Mês")
plt.xlabel("Data")
plt.ylabel("Número de Matrículas")

plt.show()
```

```python
# Checkins por horario
checkins["hour"].value_counts().sort_index().plot(kind="bar")

plt.title("Check-ins por Horário")
plt.xlabel("Hora")
plt.ylabel("Quantidade")

plt.show()
```

```python
# Time line mensal
timeline = pd.date_range(start="2018-01-01", end="2025-01-01", freq="M")

metrics = pd.DataFrame({"month": timeline})
```

```python
# Novos alunos por mes
members["join_month"] = members["join_date"].dt.to_period("M")

new_members = members.groupby("join_month").size()

metrics["new_members"] = metrics["month"].dt.to_period("M").map(new_members)
metrics["new_members"] = metrics["new_members"].fillna(0)
```

```python
# Cancelamentos por mes
members["cancel_month"] = members["cancel_date"].dt.to_period("M")

cancellations = members.groupby("cancel_month").size()

metrics["cancellations"] = metrics["month"].dt.to_period("M").map(cancellations)
metrics["cancellations"] = metrics["cancellations"].fillna(0)
```

```python
# Crescimento liquido
metrics["net_growth"] = metrics["new_members"] - metrics["cancellations"]
```

```python
# Alunos ativos por mes
active_members = []

for month in metrics["month"]:

    active = members[
        (members["join_date"] <= month) &
        ((members["cancel_date"].isna()) | (members["cancel_date"] > month))
    ].shape[0]

    active_members.append(active)

metrics["active_members"] = active_members
```

```python
# Grafico alunos ativos

plt.plot(metrics["month"], metrics["active_members"])

plt.title("Alunos Ativos ao Longo do Tempo")
plt.xlabel("Ano")
plt.ylabel("Número de Alunos")

plt.show()
```

```python
# Novos alunos vs cancelamentos
plt.plot(metrics["month"], metrics["new_members"], label="Novos alunos")
plt.plot(metrics["month"], metrics["cancellations"], label="Cancelamentos")

plt.legend()
plt.title("Novos Alunos vs Cancelamentos")

plt.show()
```

```python
# Receita mensal
payments["payment_month"] = payments["date"].dt.to_period("M")

monthly_revenue = payments.groupby("payment_month")["amount"].sum()

metrics["revenue"] = metrics["month"].dt.to_period("M").map(monthly_revenue)
metrics["revenue"] = metrics["revenue"].fillna(0)
```

```python
# Grafico de receita
plt.plot(metrics["month"], metrics["revenue"])

plt.title("Receita Mensal da Academia")
plt.xlabel("Ano")
plt.ylabel("Receita")

plt.show()
```

```python
# Ticket medio
metrics["avg_ticket"] = metrics["revenue"] / metrics["active_members"]
```

```python
# Lifetime Value
member_revenue = payments.groupby("member_id")["amount"].sum()

ltv = member_revenue.mean()

print("Lifetime Value médio:", round(ltv,2))
```

```python
# Tempo medio de permanencia
avg_membership_months = members["membership_days"].mean() / 30

print("Tempo médio de permanência:", round(avg_membership_months,2), "meses")
```

```python
# Distribuicao de permanencia
plt.hist(members["membership_days"]/30, bins=20)

plt.title("Distribuição do Tempo de Permanência")
plt.xlabel("Meses")
plt.ylabel("Número de alunos")

plt.show()
```

```python
# Quando os alunos cancelam
cancelled = members[members["churn"] == 1]

plt.hist(cancelled["membership_days"]/30, bins=20)

plt.title("Momento do Cancelamento")
plt.xlabel("Meses até cancelamento")
plt.ylabel("Número de alunos")

plt.show()
```

```python
# Curva de retencao
retention = []

months = range(1,37)

for m in months:
    
    active = (members["membership_days"]/30 >= m).mean()
    
    retention.append(active)

plt.plot(months, retention)

plt.title("Curva de Retenção de Alunos")
plt.xlabel("Meses")
plt.ylabel("Percentual de alunos ativos")

plt.show()
```

```python
# Coorte pelo mes que entraram
members["cohort"] = members["join_date"].dt.to_period("M")
members["lifetime_months"] = (members["membership_days"]/30).astype(int)
```

```python
# Tabela de retencao
cohort_data = members.groupby(["cohort","lifetime_months"]).size().unstack(fill_value=0)

cohort_sizes = cohort_data.iloc[:,0]

retention_table = cohort_data.divide(cohort_sizes, axis=0)
```

```python
# Heatmap da retencao

plt.figure(figsize=(12,6))

sns.heatmap(retention_table, cmap="Blues")

plt.title("Análise de Coorte — Retenção de Alunos")

plt.show()
```

```python
# Distribuicao da frequencia semanal
plt.hist(members["avg_weekly_visits"], bins=20)

plt.title("Distribuição de Frequência Semanal")
plt.xlabel("Visitas por semana")
plt.ylabel("Número de alunos")

plt.show()
```

```python
# Horario mais frequentado
checkins["hour"].value_counts().sort_index().plot(kind="bar")

plt.title("Check-ins por Horário")
plt.xlabel("Hora do dia")
plt.ylabel("Número de check-ins")

plt.show()
```

```python
# Frequencia por dia semanal
checkins["weekday"] = checkins["date"].dt.day_name()

checkins["weekday"].value_counts().plot(kind="bar")

plt.title("Check-ins por Dia da Semana")

plt.show()
```

```python
# Frequencia ao longo da jornada por aluno
checkins = checkins.merge(
    members[["member_id","join_date"]],
    on="member_id"
)

checkins["weeks_since_join"] = (
    (checkins["date"] - checkins["join_date"]).dt.days / 7
).astype(int)
```

```python
# Frequencia media por semana de jornada
weekly_usage = checkins.groupby("weeks_since_join").size()

weekly_usage.head(30).plot()

plt.title("Frequência ao Longo da Jornada do Aluno")
plt.xlabel("Semanas desde matrícula")
plt.ylabel("Número de check-ins")

plt.show()
```

```python
# Frequencia vs cancelamento
sns.boxplot(x="churn", y="avg_weekly_visits", data=members)

plt.title("Frequência de Treino vs Cancelamento")

plt.show()
```

```python
# Criar indicador de risco
members["low_engagement"] = members["avg_weekly_visits"] < 2
```

```python
# comparar com cancelamento
members.groupby("low_engagement")["churn"].mean()
```

```python
# Selecao variaveis para modelo
features = members[[
    "age",
    "membership_days",
    "total_checkins",
    "avg_weekly_visits"
]]

target = members["churn"]
```

```python
# Divisao treino e teste

X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.2,
    random_state=42
)
```

```python
# Modelo regressao logistica

model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
# Previsoes
predictions = model.predict(X_test)
```

```python
# Avaliacao modelo

print(classification_report(y_test, predictions))
```

```python
# Importancia das variaveis
importance = pd.DataFrame({
    "feature": features.columns,
    "importance": model.coef_[0]
})

importance.sort_values("importance", ascending=False)
```

```python
# Probabilidade de cancelamento
members["churn_probability"] = model.predict_proba(features)[:,1]
```

```python
# Lista de alunos de alto risco
high_risk = members.sort_values(
    "churn_probability",
    ascending=False
).head(20)

high_risk[[
    "member_id",
    "avg_weekly_visits",
    "membership_days",
    "churn_probability"
]]
```

```python
# KPI principais
total_members = members.shape[0]

active_members = members[members["churn"] == 0].shape[0]

churn_rate = members["churn"].mean()

avg_visits = members["avg_weekly_visits"].mean()

avg_membership = members["membership_days"].mean() / 30

print("Total de alunos:", total_members)
print("Alunos ativos:", active_members)
print("Churn rate:", round(churn_rate*100,2), "%")
print("Frequência média semanal:", round(avg_visits,2))
print("Tempo médio de permanência:", round(avg_membership,2),"meses")
```
